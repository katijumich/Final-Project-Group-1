```python
import joypy as jp
from joypy import joyplot
```

```python
df = pd.read_csv("/content/drive/MyDrive/CSC302/Final/HR-Employee-Attrition.csv")
df.head()
```

```python
%matplotlib inline
plt.figure()
plt.style.use('dark_background')
ax, fig = joyplot(
    data =df[['TrainingTimesLastYear', 'YearsAtCompany', 'PercentSalaryHike']],
    by = 'TrainingTimesLastYear', legend=True)
plt.title('Years with Company and Salary Increase by Time Training', fontsize=20, color='white', alpha=0.9)
```

```python
sns.reset_orig()
sns.swarmplot(data=df, x='WorkLifeBalance', y='YearsAtCompany', hue='Attrition', )
plt.legend(loc=1)
plt.title("Employee Loss Factoring Work Life Balance and Experience")
plt.show()
```
